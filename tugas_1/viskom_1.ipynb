```python
# import library
from PIL import Image
import math
import matplotlib.pyplot as plt
```

```python
# Memformat gambar menjadi 100 x 100

# Membuka gambar
img = Image.open('pixel_art.jpeg')

# Meresize gambar
img_resize = img.resize((100, 100))

# Menformat gambar agar RGB
img_rgb = img_resize.convert('RGB')


# Menampilkan gambar
plt.imshow(img_rgb)
plt.title('Gambar yang Telah Dipreprocessing')
plt.axis("off")
plt.show()
```

```python
# Mengubah gambar menjadi 8bit biner

# Mendapatkan nilai lebar dan tinggi gambar
widht, height = img_rgb.size

# Membuat matrix
matrix = []

# Membaca biner dalam gamabr
for i in range(height):
  row = []
  for j in range(widht):

    # Mengambil nilai rgb
    r, g, b = img_rgb.getpixel((j,i))

    # Mengkorversi setiap warna menjadi 8bit
    r_bit = f"{r:08b}"
    g_bit = f"{g:08b}"
    b_bit = f"{b:08b}"

    row.append(f"({r_bit}, {g_bit}, {b_bit})")
  matrix.append(row)


# Meyimpan matrix ke file txt
with open("nilai_8bit_rgb.txt", "w") as f:
  for i in matrix:
    f.write(" ".join(i) + "\n")

# Cetak sebagian kecil hasilnya (5x5)
for row in matrix[:5]:
    print(row[:5])

```

```python
# Ekstraksi warna per pixel

matrix_rgb = []

for i in range(height):
  row = []
  for j in range(widht):
    r, g, b = img_rgb.getpixel((j, i))

    row.append(f"({r}, {g}, {b})")
  matrix_rgb.append(row)

# Meyimpan matrix ke file txt
with open("nilai_rgb.txt", "w") as f:
  for i in matrix_rgb:
    f.write(" ".join(i) + "\n")

# Cetak sebagian kecil hasilnya (10x10)
for row in matrix_rgb[:10]:
    print(row[:10])
```

```python
# Mengubah gambar ke grayscale

#Inisiasi variabel untuk menyimpan gambar grayscale
img_gray = Image.new("L", (widht, height))

for i in range(height):
  for j in range(widht):
    r, g, b = img_rgb.getpixel((j, i))

    # rumus convert rgb to grayscale
    gray = int(0.299 * r + 0.587 * g + 0.114 * b)

    #menyimpan data pixel ke dalam variabel yang sudah diinisiasi sebelumnya
    img_gray.putpixel((j, i), gray)

#simpan gambar hasil konversi
img_gray.save('img_gray.jpg')


```

```python
# Tampilkan gambar asli dan grayscale
plt.figure(figsize=(10, 5))

# Gambar asli RGB
plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title("Gambar Asli (RGB)")
plt.axis("off")

# Gambar hasil Grayscale
plt.subplot(1, 2, 2)
gray_open = Image.open('img_gray.jpg')
plt.imshow(gray_open, cmap="gray")
plt.title("Gambar Grayscale")
plt.axis("off")

plt.show()
```

```python
# Mengekstraksi nilai warna per pixel

# Mencari ukuran pixel
widht_gray, height_gray = gray_open.size

# Inisiasi Matrix
matrix_gray = []

# Meload nilai graycale per pixel
for i in range(height_gray):
  row = []
  for j in range(widht_gray):
    gray_value = gray_open.getpixel((j,i))

    row.append(f"({gray_value})")
  matrix_gray.append(row)

# Menyimpan nilai pixel grayscale
with open("nilai_grayscale.txt", "w") as f:
  for i in matrix_gray:
      f.write(" ".join(i) + "\n")  # Gabungkan nilai dalam satu baris

# Menampilkan sebagian kecil hasilnya (10x10)
for i in matrix_gray[:10]:
  print(i[:10])
```

```python
# Mengubah gambar menjadi 1 bit alias hitam dan putih

# Inisiasi variabel threshold
threshold = 127

# inisiasi variabel untuk menyimpan nilai pixel gambar hitam putih
img_bw = Image.new("1", (widht_gray, height_gray))

for i in range(height_gray):
  for j in range(widht_gray):

    # Mengambil nilai per pixel pada gambar grayscale
    gray_value = gray_open.getpixel((j, i))

    # Menentukan apakah nilai dari gray value putih apa hitam
    bw_value = 1 if gray_value > threshold else 0

    # Menyimpan nilai ke variabel yang sudah diinisasikan
    img_bw.putpixel((j, i), bw_value)

#simpan gambar hasil konversi
img_bw.save('img_bw.jpg')
```

```python
# Tampilkan gambar asli dan hitam putih
plt.figure(figsize=(10, 5))

# Gambar asli RGB
plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title("Gambar Asli (RGB)")
plt.axis("off")

# Gambar hasil Hitam Putih
plt.subplot(1, 2, 2)
bw_open = Image.open('img_bw.jpg')
plt.imshow(bw_open, cmap="gray")
plt.title("Gambar Hitam Putih")
plt.axis("off")

plt.show()
```

```python
# Mengekstraksi nilai warna per pixel

# Mencari ukuran pixel
widht_bw, height_bw = bw_open.size

# Inisiasi Matrix
matrix_bw = []

# Meload nilai graycale per pixel
for i in range(height_bw):
  row = []
  for j in range(widht_bw):
    bw_value = bw_open.getpixel((j,i))

    row.append(f"({bw_value})")
  matrix_bw.append(row)

# Menyimpan nilai pixel grayscale
with open("nilai_hitam_putih.txt", "w") as f:
  for i in matrix_bw:
      f.write(" ".join(i) + "\n")  # Gabungkan nilai dalam satu baris

# Menampilkan sebagian kecil hasilnya (10x10)
for i in matrix_bw[:10]:
  print(i[:10])
```

```python
# Merotasikan gambar

# Menentukan titik pusat
x_center, y_center = widht // 2, height // 2

# Konversi sudut ke radian
theta = math.radians(90)

# Buat gambar baru dengan latar belakang hitam (RGB)
rotated_img = Image.new("RGB", (widht, height), (0, 0, 0))

# Rotasi per pixel
for y in range(height):
  for x in range(widht):

    # Geser ke pusat koordinat
    x_shifted = x - x_center
    y_shifted = y - y_center

    # Terapkan transformasi rotasi
    x_new = int(x_shifted * math.cos(theta) - y_shifted * math.sin(theta) + x_center)
    y_new = int(x_shifted * math.sin(theta) + y_shifted * math.cos(theta) + y_center)
    # Cek apakah dalam batas gambar baru
    if 0 <= x_new < widht and 0 <= y_new < height:
      pixel = img_rgb.getpixel((x, y))
      rotated_img.putpixel((x_new, y_new), pixel)


#simpan gambar hasil konversi
rotated_img.save('img_rotated.jpg')
```

```python
# Tampilkan gambar asli dan hasil rotasi
plt.figure(figsize=(10, 5))

# Gambar asli RGB
plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title("Gambar Asli (RGB)")
plt.axis("off")

# Gambar hasil Grayscale
plt.subplot(1, 2, 2)
rotate_open = Image.open('img_rotated.jpg')
plt.imshow(rotate_open)
plt.title("Gambar Hasil Rotasi")
plt.axis("off")

plt.show()
```

```python
# Ekstraksi warna per pixel

matrix_rotated = []

for i in range(height):
  row = []
  for j in range(widht):
    r, g, b = rotate_open.getpixel((j, i))

    row.append(f"({r}, {g}, {b})")
  matrix_rotated.append(row)

# Meyimpan matrix ke file txt
with open("nilai_rotasi.txt", "w") as f:
  for i in matrix_rotated:
    f.write(" ".join(i) + "\n")

# Cetak sebagian kecil hasilnya (10x10)
for row in matrix_rotated[:10]:
    print(row[:10])
```
